<h1 style="background-color:lightblue;padding:2rem;border-radius:5px;">How to display a radar with Matplotlib</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

#Global graphics settings
plt.style.use('dark_background')
plt.rcParams.update({'text.color' : "white",'xtick.color' : "white",'ytick.color' : "white"})

categories = ['Food Quality', 'Food Variety', 'Service Quality', 'Ambiance', 'Affordability',"Accessibility","opening hours"]

restaurant_1 = [4, 4, 5, 4, 3, 0, 0]
restaurant_2 = [5, 5, 4, 2, 2, 1, 1]
restaurant_3 = [0, 0, 5, 3, 5, 2, 2]


##### DO NOT FORGET TO PUT AT THE END OF THE LIST THE FIRST VALUE OF EACH LIST AND CATEGORIE SO THAT THE LOOP IS COMPLETED
categories.append(categories[0])
restaurant_1.append(restaurant_1[0])
restaurant_2.append(restaurant_2[0])
restaurant_3.append(restaurant_3[0])

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(restaurant_1))

plt.figure(figsize=(8, 8))
plt.subplot(polar=True)
plt.plot(label_loc, restaurant_1, label='Restaurant 1',c='red')
plt.plot(label_loc, restaurant_2, label='Restaurant 2',c='green')
plt.plot(label_loc, restaurant_3, label='Restaurant 3',c='blue')
plt.title('\nRestaurant comparison\n', size=20)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.15))
plt.show()
```
