<h1 style="background-color:lightblue;padding:2rem;border-radius:5px;">How to display a time chart with the dates</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bitcoin = pd.read_csv("https://raw.githubusercontent.com/eric2mangel/BDD/main/BTC-EUR.csv",index_col="Date",parse_dates=True)
```

<h2 style="background-color:lightgrey;padding:2rem;border-radius:5px;">Simple display of the time chart</h2>

```python
bitcoin['Close'].plot(figsize=(9,6))
plt.title('Bitcoin/EUR quotation price\n')
plt.show()
```

<h2 style="background-color:lightgrey;padding:2rem;border-radius:5px;">Centering on a period and colouring by month</h2>

```python
bitcoin['2016':'2019']['Close'].resample('M').plot(figsize=(9,5))
plt.title('Bitcoin/EUR quotation price\n')
plt.show()
```

<h3 style="background-color:lightgrey;padding:2rem;border-radius:5px;">This can then be easily smoothed by month</h3>

```python
bitcoin['2016':'2019']['Close'].resample('M').mean().plot(figsize=(9,5))
plt.title('Bitcoin/EUR quotation price\n')
plt.show()
```

<h3 style="background-color:lightgrey;padding:2rem;border-radius:5px;">And here is the chart with the moving averages</h3>

```python
plt.figure(figsize=(9,5))
plt.title('Bitcoin/EUR quotation price\n')
bitcoin.loc['2016':'2019','Close'].plot(c='lightgrey')
bitcoin.loc['2016':'2019','Close'].rolling(window=30).mean().plot(c='blue')
bitcoin.loc['2016':'2019','Close'].rolling(window=210).mean().plot(c='r')
#With the option center:
bitcoin.loc['2016':'2019','Close'].rolling(window=180, center=True).mean().plot(c='orange')
bitcoin.loc['2016':'2019','Close'].rolling(window=365, center=True).mean().plot(c='g')
plt.show()
```
